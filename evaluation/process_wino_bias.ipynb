```python
def join_tokens(tokens):
    res = ""

    for word in tokens:
        if res == "":
            res += word
        else:
            if word[0].isalnum():
                res += " " + word
            else:
                res += word
    return res
```

```python
def get_single_sample(row):
    tokens = row['tokens']
    coreference_clusters = [int(x) for x in row['coreference_clusters']]

    text = join_tokens(tokens)

    label = " ".join([tokens[coreference_clusters[0]], tokens[coreference_clusters[1]]]).lower()
    question_token1 = tokens[coreference_clusters[2]]

    sample = {"text": text, "label": label, "question_token1": question_token1, "question1": f"Who does '{question_token1}' refer to?", "question_token2": None, "question2": None}

    if len(coreference_clusters) == 6:
        question_token2 = tokens[coreference_clusters[4]]
        sample["question_token2"] = question_token2
        sample["question2"] = f"Who does '{question_token2}' refer to?"

    return sample
```

```python
type1_anti_val_processed = type1_anti_val_df.apply(lambda row: get_single_sample(row), axis=1)
type1_anti_test_processed = type1_anti_test_df.apply(lambda row: get_single_sample(row), axis=1)
type1_pro_val_processed = type1_pro_val_df.apply(lambda row: get_single_sample(row), axis=1)
type1_pro_test_processed = type1_pro_test_df.apply(lambda row: get_single_sample(row), axis=1)
```
